# **3. Anscombe**

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('anscombe')
df.head()
```

## *1. Analysez la relation entre x et y.*

```python
sns.scatterplot(x='x', y='y', data=df, color='#B649FF') 
```

## *2. Y a-t-il une meilleure manière d’analyser les choses ? Adaptez votre graphique*

```python
sns.scatterplot(x='x', y='y', hue='dataset', data=df)
```

## *3. Analysez maintenant les statistiques de chaque "paquet". Que constatez-vous ?*

```python
descriptive_stats = df.groupby('dataset').describe()
display(descriptive_stats)

fig, axes = plt.subplots(2, 2, figsize=(10, 10))

datasets = ['I', 'II', 'III', 'IV']

for i, dataset in enumerate(datasets):
    ax = axes[i // 2, i % 2]
    sns.scatterplot(data=df[df['dataset'] == dataset], x='x', y='y', ax=ax, color='#B649FF')
    ax.set_title(f'Dataset {dataset}')

plt.tight_layout()
plt.show()
```

## Observations:

- Nous observons que les graphiques sont extrêmement différents les uns des autres et pourtant bcp de valeurs de statistiques descriptives telles que la moyenne ou l'écart type sont les mêmes pour x et y et pour x les écarts interquartiles aussi pour les 3 premiers datasets.
